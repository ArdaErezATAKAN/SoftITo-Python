### 🔸 Soru 12:
Bir öğrenci listesi var. Her öğrenci `(isim, not)` tuple'ı olarak saklanıyor.
- Notu 70'in üzerinde olanları filtreleyin
- En yüksek notu alan öğrenciyi bulun
- Notların ortalamasını hesaplayın

```python
# Cevabınızı buraya yazın:
ogrenciler = [
    ("Ahmet", 85),
    ("Büşra", 62),
    ("Can", 91),
    ("Deniz", 74),
    ("Elif", 58),
    ("Furkan", 88)
]

notu_70_ustu = [isim for isim, notu in ogrenciler if notu > 70]
print("Notu 70'ten yüksek olan öğrenciler:", notu_70_ustu)

en_yuksek_ogrenci = max(ogrenciler, key=lambda x: x[1])
print(f"En yüksek notu alan öğrenci: {en_yuksek_ogrenci[0]} ({en_yuksek_ogrenci[1]})")

notlar = [notu for isim, notu in ogrenciler]
ortalama = sum(notlar) / len(notlar)
print(f"Öğrencilerin not ortalaması: {ortalama:.2f}")
```

### 🔸 Soru 13 (Set):
İki sınıfın öğrenci listesi var. `set` kullanarak:
- Her iki sınıfta da bulunan öğrencileri bulun (kesişim)
- Toplam kaç farklı öğrenci olduğunu bulun (birleşim)
- Sadece A sınıfında olup B'de olmayan öğrencileri bulun (fark)

```python
# Cevabınızı buraya yazın:
sinif_a = {"Ali", "Ayşe", "Can", "Dilan", "Ece"}
sinif_b = {"Can", "Dilan", "Fatih", "Gökçe", "Hakan"}

ortak_ogrenciler = sinif_a.intersection(sinif_b)
print("Her iki sınıfta da olanlar (Kesişim):", ortak_ogrenciler)

tum_ogrenciler = sinif_a.union(sinif_b)
print("Tüm benzersiz öğrenciler (Birleşim):", tum_ogrenciler)
print("Toplam farklı öğrenci sayısı:", len(tum_ogrenciler))

sadece_a = sinif_a.difference(sinif_b)
print("Sadece A sınıfında olanlar (Fark):", sadece_a)
```

### 🔸 Soru 14:
Bir cümledeki her harfin kaç kez geçtiğini sayan bir sözlük oluşturun (boşlukları saymayin, büyük/küçük harf farkı olmasın).

```python
# Cevabınızı buraya yazın:
cumle = "merhaba dünya"

harf_sayilari = {}

for harf in cumle.lower():
    if harf != " ":
        if harf in harf_sayilari:
            harf_sayilari[harf] += 1
        else:
            harf_sayilari[harf] = 1

print("Geçen harfler:")
for harf, sayi in harf_sayilari.items():
    print(f"'{harf}': {sayi} kez")
```

### 🔸 Soru 15:
Özyinelemeli (recursive) bir fonksiyon yazın: `faktoriyel(n)` - n! hesaplasın.
- 0! = 1
- n! = n × (n-1)!

5!, 10! değerlerini hesaplayın.

```python
# Cevabınızı buraya yazın:
def faktoriyel(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * faktoriyel(n - 1)

sonuc_5 = faktoriyel(5)
sonuc_10 = faktoriyel(10)

print(f"5! = {sonuc_5}")
print(f"10! = {sonuc_10}")
```

### 🔸 Soru 16:
`lambda` fonksiyonu kullanarak:
- Bir sayının karesini hesaplayan lambda
- İki sayının büyüğünü döndüren lambda
- `sorted()` ile bir isim listesini son harfe göre sıralayan lambda

```python
# Cevabınızı buraya yazın:
isimler = ["Zeynep", "Ali", "Mahmut", "Su", "Berk"]

karesi = lambda x: x ** 2

buyuk_olan = lambda a, b: a if a > b else b

son_harfe_gore_sirali = sorted(isimler, key=lambda x: x[-1])

print("Bir sayının karesini hesaplayan lambda:", karesi(6))
print("İki sayının büyüğünü döndüren lambda:", buyuk_olan(10, 24))
print("İsim listesini son harfe göre sıralayan lambda:", son_harfe_gore_sirali)
```

### 🔸 Soru 18 (Bütünleştirici):
Basit bir **öğrenci not sistemi** yazın:

- `ogrenciler` adında boş bir sözlük oluşturun (anahtar: isim, değer: notlar listesi)
- `ogrenci_ekle(isim)`: Yeni öğrenci ekler
- `not_ekle(isim, not_)`: Öğrenciye not ekler. Öğrenci yoksa hata mesajı verir.
- `ortalama_hesapla(isim)`: Öğrencinin not ortalamasını döndürür
- `sinif_raporu()`: Tüm öğrencileri ve ortalamalarını yazdırır, sınıf ortalamasını gösterir

En az 3 öğrenci ve her birine 3 not girerek test edin.

```python
# Cevabınızı buraya yazın:
ogrenciler = {}

def ogrenci_ekle(isim):
    ogrenciler[isim] = []

def not_ekle(isim, not_):
    if isim in ogrenciler:
        ogrenciler[isim].append(not_)
    else:
        print(f"Hata: '{isim}' adında bir öğrenci sistemde bulunamadı!")

def ortalama_hesapla(isim):
    if isim in ogrenciler and ogrenciler[isim]:
        notlar = ogrenciler[isim]
        return sum(notlar) / len(notlar)
    return 0

def sinif_raporu():
    print("--- Sınıf Raporu ---")
    toplam_ortalamalar = 0
    ogrenci_sayisi = len(ogrenciler)
    for isim, notlar in ogrenciler.items():
        ort = ortalama_hesapla(isim)
        toplam_ortalamalar += ort
        print(f"Öğrenci: {isim} | Notları: {notlar} | Ortalaması: {ort:.2f}")

    print("-" * 20)
    if ogrenci_sayisi > 0:
        genel_ortalama = toplam_ortalamalar / ogrenci_sayisi
        print(f"Sınıf Genel Ortalaması: {genel_ortalama:.2f}")
    else:
        print("Sistemde kayıtlı öğrenci yok.")

ogrenci_ekle("Ahmet")
ogrenci_ekle("Büşra")
ogrenci_ekle("Can")

not_ekle("Ahmet", 85)
not_ekle("Ahmet", 90)
not_ekle("Ahmet", 78)

not_ekle("Büşra", 92)
not_ekle("Büşra", 88)
not_ekle("Büşra", 95)

not_ekle("Can", 70)
not_ekle("Can", 65)
not_ekle("Can", 73)

sinif_raporu()
```
